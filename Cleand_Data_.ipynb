```python
# Imputing missing values in Review Rating column with the median rating of the product category

df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
## Renaming columns according to snake casing for better readability and documentation
df.columns = df.columns.str.lower().str.replace(' ' , '_')

df = df.rename(columns={'purchase_amount_(usd)' :'purchase_amount'})
```

```python
df.columns
```

```python
#purchase_amount  in purchase_amount amount is in USD Converting its into indian rupees 

df['purchase_amount'] = df['purchase_amount']*88
```

```python
df.head(10)
```

```python
## create a new column age_group



label = ["Young adult","Adult","Midle Age" , "senior"]
df['age_group'] = pd.qcut(df['age'] , q=4 , labels = label)
```

```python
df[['age' ,'age_group' ]].head(10)
```

```python
df['frequency_of_purchases']
```

```python
## create new column purchase_frequency_days

frequency_mapping = {
      'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}



df['purchase_Frequency_Days'] = df['frequency_of_purchases'].map(frequency_mapping)



 
```

## Connecting Python script to SQL-Server

```python
from sqlalchemy import create_engine

# Step 1: Connect to SQL Server using Windows Authentication
server = r".\MSSQLSERVER01"       # from your screenshot
database = "customer_behaviorr"    # your database name

# SQLAlchemy engine for Windows Authentication
engine = create_engine(
    f"mssql+pyodbc://@{server}/{database}?driver=ODBC+Driver+17+for+SQL+Server&trusted_connection=yes"
)

# Step 2: Load DataFrame into SQL Server
table_name = "customer_behavior"

df.to_sql(table_name, engine, if_exists="replace", index=False)

print(f"Data loaded into table '{table_name}' in database '{database}'.")
```
